```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
covid = pd.read_csv('owid-covid-data (1).csv')
```

```python
covid['total_cases'] = pd.to_numeric(covid['total_cases'], errors='coerce')
covid['total_deaths'] = pd.to_numeric(covid['total_deaths'], errors='coerce')

```

```python
#convert date to datetime
covid['date'] = pd.to_datetime(covid['date'])
```

```python
covid['prev_day_confirmed'] = covid.groupby('location')['total_cases'].shift(1)
covid['new_cases'] = covid['total_cases'] - covid['prev_day_confirmed']
covid['growth_rate'] = (covid['new_cases'] / covid['prev_day_confirmed']) * 100

```

```python
# Plotting
plt.figure(figsize=(14, 7))
sns.lineplot(data=covid, x='date', y='new_cases', hue='location')
plt.title('Daily New Cases Over Time by Country')
plt.xlabel('Date')
plt.ylabel('Daily New Cases')
plt.legend(title='location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
# Plotting
plt.figure(figsize=(14, 7))
sns.lineplot(data=covid, x='date', y='total_deaths', hue='location')
plt.title('Recovery Rate Over Time by Country')
plt.xlabel('Date')
plt.ylabel('Recovery Rate (%)')
plt.legend(title='location', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python

heatmap = covid.pivot_table(index="date", columns="location", values="total_cases", aggfunc='max')


plt.figure(figsize=(14, 10))
sns.heatmap(heatmap, cmap="YlGnBu", linewidths=.5)
plt.title('Heatmap of COVID-19 Confirmed Cases by Country and Date')
plt.show()

```

```python
total_cases_by_location = covid.groupby('continent')['total_cases'].max().reset_index()
```

```python

# Plot bar chart
plt.figure(figsize=(14, 7))
sns.barplot(data=total_cases_by_location, x='continent', y='total_cases')
plt.title('Total Confirmed Cases by Country')
plt.xlabel('Country')
plt.ylabel('Total Confirmed Cases')
plt.xticks(rotation=90)
plt.show()
```

```python
covid['new_cases'] = covid['total_cases'].diff().fillna(0)
```

```python
plt.figure(figsize=(14, 7))
sns.histplot(covid['new_cases'], bins=50, kde=True)
plt.title('Distribution of Daily New Cases')
plt.xlabel('Daily New Cases')
plt.ylabel('Frequency')
plt.show()
```
